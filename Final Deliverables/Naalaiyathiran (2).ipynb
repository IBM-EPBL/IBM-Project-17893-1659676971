```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("test_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
df[['Bed Grade']].value_counts()
```

```python
df
```

```python

df.replace({'Type of Admission':{'Emergency':3,'Trauma':1,'Urgent':2}}, inplace = True)
df.replace({'Department':{'gynecology':3,'anesthesia':1,'radiotherapy':2,'TB & Chest disease':4,'surgery':5}}, inplace = True)
```

```python
df.replace({'Severity of Illness':{'Minor':1,'Moderate':2,'Extreme':3}}, inplace = True)
y = df['Admission_Deposit']
```

```python
x = df[['Severity of Illness']]
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size = 0.3 )
```

```python
from sklearn.linear_model import LinearRegression
xtrain
```

```python
lr = LinearRegression()
lr.fit(xtrain , ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
plt.scatter(xtest,ypred)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
```

```python
mean_squared_error(ytest, ypred)
```

```python
mean_absolute_error(ytest, ypred)
```

```python
r2_score(ytest, ypred)
```
